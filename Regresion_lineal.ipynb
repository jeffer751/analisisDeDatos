```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("C:/Users/Stella Cuero/OneDrive/Escritorio/STEVEN/universidad/Analisis de datos/UNIDAD 2/regresion_lineal/data.csv", sep=",")
```

```python
data
```

```python
data.plot.scatter(x="metro" , y="precio")
plt.show()
```

```python
#Regresion lineal  
regresion = linear_model.LinearRegression()
```

```python
#Array 
metro = data["metro"].values.reshape((-1,1))
```

```python
#Modelo
modelo = regresion.fit(metro, data["precio"])
print("Interseccion (b)", modelo.intercept_)
print("Pendiente (m)", modelo.coef_)
```

```python
entrada= [[5],[15],[20],[25]]
predicciones = modelo.predict(entrada)
print(predicciones)
```

```python
data.plot.scatter(x="metro", y="precio", label='Dato inicial')
plt.scatter(entrada, predicciones, color='yellow')
plt.plot(entrada, predicciones, color='red', label ='Linea de regresión')
plt.xlabel('metro')
plt.ylabel('precio')
plt.legend()
plt.show()       

```
